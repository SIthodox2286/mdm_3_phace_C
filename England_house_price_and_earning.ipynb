```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import openpyxl 

# Load Data Functions
from load_data import load_house_price_and_earning_by_district
```

```python
median_house_price, lower_quatile_house_price, median_earning, lower_quatile_earning = load_house_price_and_earning_by_district()
```

```python
lower_quatile_earning.head()
```

```python
la_names = lower_quatile_earning.loc[:,'Local authority name']
lq_earning_2020 = lower_quatile_earning.loc[:,2020]

objects = list(la_names)
y_pos = np.arange(len(objects))

fig, ax = plt.subplots()
plt.bar(y_pos, list(lq_earning_2020), align='center',width=0.8)
plt.xticks(y_pos, objects)
figure = plt.gcf()
figure.set_size_inches(80,5)
plt.setp(ax.get_xticklabels(), fontsize=10, rotation='vertical')
plt.ylabel('Earning(Pounds)')
plt.title('Lower Quatile Earning by Local Authority Districts(England)')

plt.show()
```
